## Actividades del Taller



### 1. Diga qué índices serían pertinentes tener y póngalos.

Considero pertinentes los índices año, mes, día y hora

```python
#variables como indice del dataframe
data_pm25_multi=pd.read_csv("PRSA_data_2010.csv", index_col=[1, 2, 3, 4])
display(data_pm25_multi)
```

### 2. Elimine registros que tengan NA en la columna "pm2.5" 

```python
#eliminar valores (se eliminan 2067 filas)
data_pm25.dropna(subset=["pm2.5"], axis=0, inplace=False)
```

```python
#nueva data sin NA
pm25_no_na = data_pm25[data_pm25["pm2.5"].notna()]
print(data_pm25.shape)
print(pm25_no_na.shape)
```

### 3. Realice un describe de todas las variables por año. Además de realizar un gráfico como boxplot con el que pueda comparar los resultados dados. 


```python
#describe de todas las variables por año
print(pm25_no_na.groupby('year').describe(include="all"))
```

```python
#gráfico boxplot
pm25_no_na.boxplot(column =['pm2.5'], by ='year',  grid = False, figsize=(8,6)) 
plt.show()
```

### 4. Para finalizar, agrupe por año el valor de la variable "pm2.5", calcule el promedio y ordene la base de datos de manera ascendente por el promedio

```python
#promedio de la variable pm2.5 agrupada por año y ordenada de manera ascendente
pm25_no_na_sort=(pm25_no_na.groupby('year')[['pm2.5']]
                .mean()
                .sort_values('pm2.5'))
display(pm25_no_na_sort)
```
